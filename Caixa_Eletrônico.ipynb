Caixa Eletrônico - Lógica de Programação

```python
cem = int(input("Qual a quantidade inicial de notas de cem?\n")) # armazena a quantidade inicial de cada tipo de nota
cinq = int(input("Qual a quantidade inicial de notas de cinquenta?\n"))
dez = int(input("Qual a quantidade inicial de notas de dez?\n"))
cin = int(input("Qual a quantidade inicial de notas de cinco?\n"))
um = int(input("Qual a quantidade inicial de notas de um?\n"))
total = cem*100 + cinq*50 + dez*10 + cin*5 + um*1 # gera o valor total
print(f"Situação atual: {cem} {cinq} {dez} {cin} {um} {total}\n") # mostra a situação atual
n = int(input("Quantas operações serão realizadas?\n")) # guarda o número de operações
```

```python
i = 1
while i <= n: # loop que fará cada uma das n operações
    ope = int(input("Digite 0 se a operação será um saque ou digite 1 se a operação será um depósito:\n")) # guarda o valor correspondente à operação que será realizada
    if ope==0: # comandos a serem realizados no caso de um saque
        saque = int(input("Qual o valor do saque a ser realizado?\n")) # guarda o valor do saque
        count100 = 0 # conta o número de notas de 100 que estão sendo sacadas
        count50 = 0 # conta o número de notas de 50 que estão sendo sacadas
        count10 = 0 # conta o número de notas de 10 que estão sendo sacadas
        count5 = 0 # conta o número de notas de 5 que estão sendo sacadas
        count1 = 0 # conta o número de notas de 1 que estão sendo sacadas
        saqueint = saque # armazena o valor do saque, uma variável auxiliar
        while saque >= 100 and cem > 0: # verifica se o saque possui um valor maior ou igual a 100 e se há notas de 100 para realizar a operação
            cem = cem - 1
            saque = saque - 100
            count100 = count100 + 1
        while saque >= 50 and cinq > 0: # verifica se o saque possui um valor maior ou igual a 50 e se há notas de 50 para realizar a operação
            cinq = cinq - 1
            saque = saque - 50
            count50 = count50 + 1
        while saque >= 10 and dez > 0: # verifica se o saque possui um valor maior ou igual a 10 e se há notas de 10 para realizar a operação
            dez = dez - 1
            saque = saque - 10
            count10 = count10 + 1
        while saque >= 5 and cin > 0: # verifica se o saque possui um valor maior ou igual a 5 e se há notas de 5 para realizar a operação
            cin = cin - 1
            saque = saque - 5
            count5 = count5 + 1
        while saque >= 1 and um > 0: # verifica se o saque possui um valor maior ou igual a 1 e se há notas de 1 para realizar a operação
            um = um - 1
            saque = saque - 1
            count1 = count1 + 1
        if saque==0: # se o saque foi realizado com sucesso
            total = cem*100 + cinq*50 + dez*10 + cin*5 + um*1
            print(f"O saque de {saqueint} ducas foi efetuado!")
            print(f"Situacao atual: {cem} {cinq} {dez} {cin} {um} {total}\n")
        if saque != 0: # se o saque não foi realizado com sucesso
            cem = cem + count100 # repõe as notas de 100 retiradas na operação
            cinq = cinq + count50 # repõe as notas de 50 retiradas na operação
            dez = dez + count10 # repõe as notas de 10 retiradas na operação
            cin = cin + count5 # repõe as notas de 5 retiradas na operação
            um = um + count1 # repõe as notas de 1 retiradas na operação
            total = cem*100 + cinq*50 + dez*10 + cin*5 + um*1
            print(f"Faltam notas para realizar o saque de {saqueint} ducas\n")
            print(f"Situacao atual: {cem} {cinq} {dez} {cin} {um} {total}\n")
    elif ope==1: # no caso de um depósito
        dep1 = int(input("Qual quantidade de notas de 100 vc deseja depositar?\n")) # guarda o número de notas de 100 depositadas
        dep2 = int(input("Qual quantidade de notas de 50 vc deseja depositar?\n")) # guarda o número de notas de 50 depositadas
        dep3 = int(input("Qual quantidade de notas de 10 vc deseja depositar?\n")) # guarda o número de notas de 10 depositadas
        dep4 = int(input("Qual quantidade de notas de 5 vc deseja depositar?\n")) # guarda o número de notas de 5 depositadas
        dep5 = int(input("Qual quantidade de notas de 1 vc deseja depositar?\n")) # guarda o número de notas de 1 depositadas
        deposito = dep1*100 + dep2*50 + dep3*10 + dep4*5 + dep5*1 # armazena o valor total do depósito
        cem = cem + dep1 # adiciona a quantidade de notas de 100 depositadas no caixa
        cinq = cinq + dep2 # adiciona a quantidade de notas de 50 depositadas no caixa
        dez = dez + dep3 # adiciona a quantidade de notas de 10 depositadas no caixa
        cin = cin + dep4 # adiciona a quantidade de notas de 5 depositadas no caixa
        um = um + dep5 # adiciona a quantidade de notas de 1 depositadas no caixa
        total = cem*100 + cinq*50 + dez*10 + cin*5 + um*1
        print(f"Depósito de {deposito} ducas efetuado!\n")
        print(f"Situacao atual: {cem} {cinq} {dez} {cin} {um} {total}\n")
    i = i + 1 # próxima operação
```
